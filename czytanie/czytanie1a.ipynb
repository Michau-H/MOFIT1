```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
import os
```

```python
metody = ['Euler', 'Verlet', 'RK4']
folder = 'results_1a'
```

```python
def czytaj(file_path):
    with open(file_path, 'r') as file:
        data = []
        for line in file:
            values = list(map(float, line.split()))
            data.append(values)
        data = np.array(data)
        dt = data[0][0]
        data = data[1:,:]
        data = data.T
        return data, dt
```

```python
def rysuj_xt(data, i, krokT) :
    t = data[i][0]
    x = data[i][1]
    plt.plot(t,x, label=krokT)

def rysuj_vt(data, i, krokT) :
    t = data[i][0]
    v = data[i][2]
    plt.plot(t,v, label=krokT)

def rysuj_vx(data, i, krokT) :
    x = data[i][1]
    v = data[i][2]
    plt.plot(x,v, label=krokT)

def rysuj_et(data, i, krokT) :
    t = data[i][0]
    e = data[i][3]
    plt.plot(t,e, label=krokT)
```

```python
def ogarnij_czas(t, dt) :
    return np.arange(0,t/dt+1) * dt
```

```python
data_Euler = [[],[],[],[],[]]
data_Verlet = [[],[],[],[],[]]
data_RK4 = [[],[],[],[],[]]

DeltaTE = np.zeros(5)
DeltaTV = np.zeros(5)
DeltaTR = np.zeros(5)
```

```python
for metoda in metody:
    for i in range(5) :
        plik = f'{metoda}_{i}.txt'
        sciezka = os.path.join(folder, plik)
        if os.path.exists(sciezka):
            print(f"Otwieranie {plik}")
            if metoda=='Euler' :
                data_Euler[i], DeltaTE[i] = czytaj(sciezka)
            if metoda=='Verlet' :
                data_Verlet[i], DeltaTV[i] = czytaj(sciezka)
            if metoda=='RK4' :
                data_RK4[i], DeltaTR[i] = czytaj(sciezka)

        else:
            print(f"nie istnieje.")
```

```python
for i in range(5) :
    data_Euler[i][0] = ogarnij_czas(100, DeltaTE[i])
for i in range(5) :
    data_Verlet[i][0] = ogarnij_czas(100, DeltaTV[i])
for i in range(5) :
    data_RK4[i][0] = ogarnij_czas(100, DeltaTR[i])
```

```python
for i in range(5):
    rysuj_xt(data_Euler, i, DeltaTE[i])
    plt.legend(loc='lower left')
plt.title('euler x(t)')
plt.show()
for i in range(5):
    rysuj_xt(data_Verlet, i, DeltaTV[i])
    plt.legend(loc='lower left')
plt.title('verlet x(t)')
plt.show()
for i in range(5):
    rysuj_xt(data_RK4, i, DeltaTR[i])
    plt.legend(loc='lower left')
plt.title('rk4 x(t)')
plt.show()
```

```python
for i in range(5):
    rysuj_vt(data_Euler, i, DeltaTE[i])
    plt.legend(loc='lower left')
plt.title('euler x(t)')
plt.show()
for i in range(5):
    rysuj_vt(data_Verlet, i, DeltaTV[i])
    plt.legend(loc='lower left')
plt.title('verlet x(t)')
plt.show()
for i in range(5):
    rysuj_vt(data_RK4, i, DeltaTR[i])
    plt.legend(loc='lower left')
plt.title('rk4 x(t)')
plt.show()
```

```python
for i in range(5):
    rysuj_vx(data_Euler, i, DeltaTE[i])
    plt.legend(loc='lower left')
plt.title('euler x(t)')
plt.show()
for i in range(5):
    rysuj_vx(data_Verlet, i, DeltaTV[i])
    plt.legend(loc='lower left')
plt.title('verlet x(t)')
plt.show()
for i in range(5):
    rysuj_vx(data_RK4, i, DeltaTR[i])
    plt.legend(loc='lower left')
plt.title('rk4 x(t)')
plt.show()
```

```python
for i in range(5):
    rysuj_et(data_Euler, i, DeltaTE[i])
    plt.legend(loc='lower left')
plt.title('euler x(t)')
plt.show()
for i in range(5):
    rysuj_et(data_Verlet, i, DeltaTV[i])
    plt.legend(loc='lower left')
plt.title('verlet x(t)')
plt.show()
for i in range(5):
    rysuj_et(data_RK4, i, DeltaTR[i])
    plt.legend(loc='lower left')
plt.title('rk4 x(t)')
plt.show()
```
