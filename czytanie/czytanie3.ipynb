```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
import os
```

```python
metody = ['Euler', 'Verlet', 'RK4']
folder = 'results_3'
```

```python
def czytaj(file_path):
    with open(file_path, 'r') as file:
        data = []
        for line in file:
            values = list(map(float, line.split()))
            data.append(values)
        data = np.array(data)
        data = data.T
        return data
```

```python
def rysuj_xt(data, i, krokT) :
    t = data[i][0]
    x = data[i][1]
    plt.plot(t,x, label=krokT)

def rysuj_vt(data, i, krokT) :
    t = data[i][0]
    v = data[i][2]
    plt.plot(t,v, label=krokT)

def rysuj_vx(data, i, krokT) :
    x = data[i][1]
    v = data[i][2]
    plt.plot(x,v, label=krokT)

def rysuj_et(data, i, krokT) :
    t = data[i][0]
    e = data[i][3]
    plt.plot(t,e, label=krokT)

def rysuj_dtt(data, i, krokT) :
    t = data[i][0][1:]
    dt = data[i][0][1:]-data[i][0][0:-1]
    # print(t.size,dt.size)
    plt.plot(t,dt, label=krokT)
```

```python
data_Euler = [[],[]]
data_Verlet = [[],[]]
data_RK4 = [[],[]]
```

```python
for metoda in metody:
    for i in range(2) :
        plik = f'{metoda}_{i}.txt'
        sciezka = os.path.join(folder, plik)
        if os.path.exists(sciezka):
            print(f"Otwieranie {plik}")
            if metoda=='Euler' :
                data_Euler[i] = czytaj(sciezka)
            if metoda=='Verlet' :
                data_Verlet[i] = czytaj(sciezka)
            if metoda=='RK4' :
                data_RK4[i] = czytaj(sciezka)

        else:
            print(f"nie istnieje.")
```

```python
for i in range(1):
    rysuj_xt(data_Euler, i, 'Euler')
    plt.legend(loc='lower left')
plt.title('x(t)')
plt.show()
# for i in range(1):
#     rysuj_xt(data_Verlet, i, 'Verlet')
#     plt.legend(loc='lower left')
# plt.title('verlet x(t)')
# plt.show()
for i in range(1):
    rysuj_xt(data_RK4, i, 'RK4')
    plt.legend(loc='lower left')
# plt.title('rk4 x(t)')
plt.show()
```

```python
for i in range(1):
    rysuj_vt(data_Euler, i, 'Euler')
    plt.legend(loc='lower left')
plt.title('v(t)')
# plt.show()
for i in range(1):
    rysuj_vt(data_Verlet, i, 'Verlet')
    plt.legend(loc='lower left')
# plt.title('verlet x(t)')
# plt.show()
for i in range(1):
    rysuj_vt(data_RK4, i, 'RK4')
    plt.legend(loc='lower left')
# plt.title('rk4 x(t)')
plt.show()
```

```python
for i in range(1):
    rysuj_et(data_Euler, i, 'Euler')
    plt.legend(loc='upper left')
plt.title('e(t)')
# plt.show()
for i in range(1):
    rysuj_et(data_Verlet, i, 'Verlet')
    plt.legend(loc='upper left')
# plt.title('verlet x(t)')
# plt.show()
for i in range(1):
    rysuj_et(data_RK4, i, 'RK4')
    plt.legend(loc='upper left')
# plt.title('rk4 x(t)')
plt.show()
```

```python
for i in range(1):
    rysuj_dtt(data_Euler, i, 'Euler')
    plt.legend(loc='upper left')
plt.title('e(t)')
# plt.show()
for i in range(1):
    rysuj_dtt(data_Verlet, i, 'Verlet')
    plt.legend(loc='upper left')
# plt.title('verlet x(t)')
# plt.show()
for i in range(1):
    rysuj_dtt(data_RK4, i, 'RK4')
    plt.legend(loc='upper left')
# plt.title('rk4 x(t)')
plt.show()
```
